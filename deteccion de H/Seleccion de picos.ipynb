### Importando espectro

```python
import pandas as pd
```

```python
data_path = 'C:/Users/gomez/Documents/LIBS/Data/Calibracion Zr2.5Nb - 4.53 J- 2.92us/ARG-1/a1.ols'
```

```python
df = pd.read_csv(data_path, delimiter = "\t", skiprows = 6)
df
```

### Grafico de inspeccion 

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
fig = plt.figure()
f1 = fig.add_subplot(1, 1, 1)
f1.plot('Wavelength', 'Counts', data=df)
f1.set_xlabel('Wavelength (nm)')
f1.set_ylabel('Intensity')
fig.set_size_inches(12, 6)
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x = df['Wavelength'], y = df['Counts'], mode = 'lines', name = 'hola'))
fig.update_layout(
        title = "Mean spectrum for each class",
        xaxis_title = "Wavelength (nm)",
        yaxis_title = "Relative intensity (-)"
)
fig.show()
```
